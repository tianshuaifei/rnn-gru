```python
import re
import cPickle
reader = open("train_set.txt",'r')
red = []
word2id={}
label2id={}
key=0
labelkey=0
while True:
    line = reader.readline()  
    if len(line) == 0:
            break
    if len(line)>9:      
        red=line.split("\t")
        if red[0] not in word2id.values(): 
            word2id.setdefault(key,red[0])
            key+=1
        if red[4].replace('\r\n','') not in label2id.values():
            label2id.setdefault(labelkey,red[4].replace('\r\n',''))
            labelkey+=1
            
reader.close()
fn = 'label2idx.pkl'
with open(fn, 'w') as f:                   
    cPickle.dump(label2id,f)
    f.close()
fn1 = 'word2idx.pkl'
with open(fn1, 'w') as f1:                   
    cPickle.dump(word2id,f1)
    f1.close()
print word2id
print label2id
```

```python
import re
import cPickle
import numpy as np
reader = open("train_set.txt",'r')
red = []
list1=[]
list2=[]
a=[]
c=[]
a1=[]
c1=[]
while True:
    line = reader.readline()
    if len(line) == 0:
            break
    if len(line)>9:      
        red=line.split("\t")
        if red[0] in word2idx.values():
            a.append( word2idx.values().index(red[0])) 
        else:
            a.append(0) 
        if red[4].replace('\r\n','') in label2idx.values():
            a1.append(label2idx.values().index(red[4].replace('\r\n','')))   
    elif len(line)>0 and len(line)<9 :
        c=a[:]
        c1=a1[:]
        if len(c)>0:
            list1.append(c)
            list2.append(c1)
        del a[:]
        del a1[:]
reader.close()
tup = (np.array(list1[:]),np.array(list2[:]))
fn = 'train_data.pkl'
with open(fn, 'w') as f:                   
    cPickle.dump(tup,f)
    f.close()      
```
